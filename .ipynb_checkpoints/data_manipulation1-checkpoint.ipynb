```python
def get_ticker(company_name):
    url = "https://query2.finance.yahoo.com/v1/finance/search"
    ua = UserAgent()
    chrome_pass = ua.chrome
#     user_agent = chrome_pass
    headers = {'User-Agent': ua.chrome}
    params = {"q": company_name, "quotes_count": 1, "country": "United States"}
    
    try:
        res = requests.get(url=url, params=params, headers=headers, timeout=10)
        res.raise_for_status()
        data = res.json()
        return data['quotes'][0]['symbol'] if data['quotes'] else None
    except Exception as e:
        print(f"Error fetching ticker for {company_name}: {e}")
        return None

#     res = requests.get(url=yfinance, params=params, headers={'User-Agent': user_agent})
#     data = res.json()

#     company_code = data['quotes'][0]['symbol']
#     return company_code
```

```python
# Use tqdm for a progress bar
for idx, row in tqdm(full_scm_df.iterrows(), total=len(full_scm_df)):
    company_name = row["Company Name"]
    
    # Only fetch if not already filled
    if pd.isna(full_scm_df.at[idx, "Ticker"]):
        ticker = get_ticker(company_name)
        full_scm_df.at[idx, "Ticker"] = ticker
        
        # Sleep a bit between requests (you can tune this)
        time.sleep(0.1)  # 1.5 seconds between calls to avoid hammering the server

        # Optional: Save progress every N rows
#         if idx % 5 == 0:
#             full_scm_df.to_csv("ticker_progress.csv", index=False)
```
