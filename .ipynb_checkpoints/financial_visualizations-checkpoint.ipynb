```python
financial_data = financial_data[[i for i in financial_data.columns.tolist() if i[0] == 'Close']]
```

```python
financial_data.columns = financial_data.columns.droplevel('Price')
```

```python
cum_rtns = financial_data.copy()
```

```python
for col in cum_rtns.columns.tolist():
    cum_rtns[f'{col}_cumrtn'] = np.cumprod(1 + cum_rtns[col].pct_change()) - 1
    
cum_rtns = cum_rtns[[i for i in cum_rtns.columns if 'cumrtn' in i]]
```

```python
cum_rtns.columns = [i[:-7] for i in cum_rtns.columns.tolist()]
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
financial_data = financial_data.loc[financial_data.index < '2021-01-01']
cum_rtns = cum_rtns.loc[cum_rtns.index < '2021-01-01']
```

```python
# Step 1: Create mapping dictionaries from `clean_top`
scm_dict = clean_top_df.set_index('Ticker')['SCM Practices'].to_dict()
eff_dict = clean_top_df.set_index('Ticker')['Operational Efficiency Score'].to_dict()
name_dict = clean_top_df.set_index('Ticker')['Company Name'].to_dict()

# Step 2: Melt the financial_data to long format
financial_long = financial_data.reset_index().melt(id_vars='Date', 
                                                   var_name='Ticker', 
                                                   value_name='Price')

# Step 3: Map SCM Practices and Operational Efficiency Score
financial_long['SCM Practices'] = financial_long['Ticker'].map(scm_dict)
financial_long['Operational Efficiency Score'] = financial_long['Ticker'].map(eff_dict)
financial_long['Company Name'] = financial_long['Ticker'].map(name_dict)

spy_data = financial_long[financial_long['Ticker'] == 'SPY']
other_data = financial_long[financial_long['Ticker'] != 'SPY']
```

```python
# Step 4: Plot
fig = px.line(
    other_data,
    x='Date',
    y='Price',
    color='Ticker',
    hover_data={
        'Ticker': True,
        'Company Name': True,
        'Price': ':.2f',
        'SCM Practices': True,
        'Operational Efficiency Score': ':.1f',
        'Date': True  # optional: hide date in hover if x-axis already shows it
    },
    title="Stock Prices Over Time with SCM Practices and Efficiency Scores"
)

fig.add_vline(
    x='2020-03-18',
    line_width=2,
    line_dash="dash",  # Options: "dash", "dot", "dashdot"
    line_color="black"
)

fig.add_annotation(
    x='2019-12-20',
    y=0.9,  # y=1 means top of the plot area (relative coordinates)
    yref='paper',
    text="COVID Market Low",
    showarrow=False,
    font=dict(color="black"),
    align="center"
)

fig.add_trace(
    go.Scatter(
        x=spy_data['Date'],
        y=spy_data['Price'],
        mode='lines',
        name='SPY',
        line=dict(color='black', dash='dot'),
        hovertemplate='<b>SPY</b><br>Date: %{x}<br>Price: %{y:.2f}<extra></extra>'
    )
)

fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='closest',
    xaxis=dict(showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showgrid=True, gridcolor='lightgrey'),
    font=dict(color='black')
)


# fig.update_layout(hovermode='x unified')  # makes hover show values across all lines at once
fig.show()
```

```python
fig.write_html("stock_prices_plot.html")
```

```python
# Step 1: Create mapping dictionaries from `clean_top`
scm_dict = clean_top_df.set_index('Ticker')['SCM Practices'].to_dict()
eff_dict = clean_top_df.set_index('Ticker')['Operational Efficiency Score'].to_dict()
name_dict = clean_top_df.set_index('Ticker')['Company Name'].to_dict()

# Step 2: Melt the financial_data to long format
financial_long2 = cum_rtns.reset_index().melt(id_vars='Date', 
                                                   var_name='Ticker', 
                                                   value_name='Cumulative Return')

# Step 3: Map SCM Practices and Operational Efficiency Score
financial_long2['SCM Practices'] = financial_long2['Ticker'].map(scm_dict)
financial_long2['Operational Efficiency Score'] = financial_long2['Ticker'].map(eff_dict)
financial_long2['Company Name'] = financial_long2['Ticker'].map(name_dict)

spy_data = financial_long2[financial_long2['Ticker'] == 'SPY']
other_data = financial_long2[financial_long2['Ticker'] != 'SPY']
```

```python
# Step 4: Plot
fig = px.line(
    other_data,
    x='Date',
    y='Cumulative Return',
    color='Ticker',
    hover_data={
        'Ticker': True,
        'Company Name': True,
        'Cumulative Return': ':.2%',
        'SCM Practices': True,
        'Operational Efficiency Score': ':.1f',
        'Date': True  # optional: hide date in hover if x-axis already shows it
    },
    title="Cumulative Return over Time of Top Companies by Operational Efficiency Score by SCM Practice"
)

fig.add_vline(
    x='2020-03-18',
    line_width=2,
    line_dash="dash",  # Options: "dash", "dot", "dashdot"
    line_color="black"
)

fig.add_annotation(
    x='2019-12-20',
    y=0.9,  # y=1 means top of the plot area (relative coordinates)
    yref='paper',
    text="COVID Market Low",
    showarrow=False,
    font=dict(color="black"),
    align="center"
)

fig.add_trace(
    go.Scatter(
        x=spy_data['Date'],
        y=spy_data['Cumulative Return'],
        mode='lines',
        name='SPY',
        line=dict(color='black', dash='dot'),
        hovertemplate='<b>SPY</b><br>Date: %{x}<br>Cumulative Return: %{y:.2%}<extra></extra>'
    )
)


fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    hovermode='closest',
    xaxis=dict(showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showgrid=True, gridcolor='lightgrey'),
    font=dict(color='black')
)


# fig.update_layout(hovermode='x unified')  # makes hover show values across all lines at once
fig.show()
```

```python
fig.write_html("cum_rtns_plot.html")
```
